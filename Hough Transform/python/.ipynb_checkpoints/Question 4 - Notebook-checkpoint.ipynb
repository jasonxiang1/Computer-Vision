set up basis to import example image file to work with

```python
img = cv2.imread('../data/img01.jpg')
img_rgb = np.asarray(img)

img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img_gray = np.asarray(img_gray)

#print out image file
plt.imshow(img_gray,cmap="gray");plt.show()
```

### acquire edge image

```python
#declare sigma value
sigma = 1.5
```

```python
Img1, Io, Ix, Iy = myEdgeFilter(img_gray, sigma)

#apply thresholding to image
Thresh = 0.2
InnputImage = 1.0 * (Img1 > Thresh)

#output Im edge image
plt.imshow(InnputImage,cmap="gray");plt.show()
print(np.amax(InnputImage))
print(np.amin(InnputImage))
```

```python
#declare variables of Img1
Img1_rows = len(Img1)
Img1_cols = len(Img1[0])
Img1_val_mat = np.array([Img1_rows, Img1_cols])
print(Img1_val_mat)
rho_max = np.int_(np.ceil(np.sqrt(np.dot(np.transpose(Img1_val_mat),Img1_val_mat))))
print(rho_max)
print(rho_max.dtype)

#declare rhores and thetares
rhoRes = 1
thetaRes = 1
```

```python
#set theta
theta_max = 360
theta = np.arange(0,theta_max,thetaRes)
print(theta.shape)
```

```python
#set rho
rho = np.arange(0,rho_max,rhoRes)
print(rho.shape)
```

```python
#calculate total length of theta and rho based on resolutions params
theta_vec_len = np.int(np.ceil(theta_max*(1/thetaRes)))
rho_vec_len = np.int(np.ceil((rho_max)*(1/rhoRes)))

#declare accumulator matrix as zero matrix that is theta x rho
accum_mat = np.zeros((theta_vec_len,rho_vec_len),dtype=int)
print(accum_mat.shape)
```

```python
#convert image into 1D array
Img1_reshaped_1D = Img1.reshape((Img1_rows*Img1_cols,1))
print(Img1_reshaped_1D.shape)
```

```python
#create x vector
row_Img1_xvec_init = np.arange((Img1_cols),dtype=int)
print(row_Img1_xvec_init.size)
row_Img1_xvec_repeat = np.tile(row_Img1_xvec_init,Img1_rows)
print(row_Img1_xvec_repeat.size)
print(row_Img1_xvec_repeat[600:650])
```

```python
#create y vector
row_Img1_yvec_init = np.arange((Img1_rows),dtype=int)
print(row_Img1_yvec_init.size)
row_Img1_yvec_repeat = np.repeat(row_Img1_yvec_init,Img1_cols)
print(row_Img1_yvec_repeat.size)
print(row_Img1_yvec_repeat[600:650])
```

```python
#combine 1D x and 1D y
Img1_vectored = np.concatenate((row_Img1_xvec_repeat,row_Img1_yvec_repeat))
print(Img1_vectored.shape)
Img1_vectored_2D_horiz = Img1_vectored.reshape((2,np.int(Img1_vectored.size/2)))
print(Img1_vectored_2D_horiz.shape)
print(Img1_vectored_2D_horiz[:,620:660])
Img1_vectored_2D_vert = np.transpose(Img1_vectored_2D_horiz)
print(Img1_vectored_2D_vert[620:660,:])
```

```python
#create vector of cos and sin, respectively
cos_vectored = np.cos(np.deg2rad(theta))
sin_vectored = np.sin(np.deg2rad(theta))

#combine cos and sin vectors
theta_cos_sin = np.concatenate((cos_vectored,sin_vectored))
theta_cos_sin_reshape = theta_cos_sin.reshape((2,np.int(len(theta_cos_sin)/2)))
print(theta_cos_sin_reshape.shape)
```

```python
#multiple together
rho_vectored = np.dot(Img1_vectored_2D_vert,theta_cos_sin_reshape)
print(rho_vectored.shape)
```

```python
#reshape rho_vectored back to img dimensions and multiply with image
rho_vectored_image_shape = rho_vectored.reshape(Img1_rows,Img1_cols,theta_max)
print(rho_vectored_image_shape[220,330,:])
print(np.amax(rho_vectored_image_shape))
```

```python
rho_on_image = rho_vectored_image_shape * InnputImage
print(np.amax(rho_on_image))
print(np.amin(rho_on_image))
print(rho_on_image.shape)



rho_on_image[rho_on_image<0] = -1 * rho_on_image[rho_on_image<0]

print(np.amax(rho_on_image))
print(np.amin(rho_on_image))
print(rho_on_image.shape)

rho_on_image_rounded = np.int_(np.around(rho_on_image))
print(np.amax(rho_on_image_rounded))
print(np.amin(rho_on_image_rounded))
```

```python
print(accum_mat.shape)
```

```python
for i in range(0,len(theta)):
    unique, counts = np.unique(rho_on_image_rounded[:,:,i],return_counts=True)
    accum_mat[i,unique-1] = accum_mat[i,unique-1] + counts
    
accum_mat[:,0] = 0
accum_mat[:,-1] = 0
print(np.amax(accum_mat))
print(np.amin(accum_mat))

print(accum_mat)
plt.imshow(np.transpose(accum_mat));plt.show()
```

```python
unique, counts = np.unique(rho_on_image_rounded[0,:,0],return_counts=True)
print(counts)
print(range(0,len(theta)))
```
