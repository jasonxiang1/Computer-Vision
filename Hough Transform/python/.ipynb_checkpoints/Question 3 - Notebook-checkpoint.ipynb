# Import Packages

```python
import cv2
import numpy as np
import numpy.matlib as npm
import math

from GaussianKernel import Gauss2D
from myImageFilter import myImageFilter


#module to delete
from matplotlib import pyplot as plt
import matplotlib
```

# Import Sample Image File

```python
img = cv2.imread('../data/img01.jpg')
img_rgb = np.asarray(img)

img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img_gray = np.asarray(img_gray)

#print out image file
plt.imshow(img_gray,cmap="gray");plt.show()
```

```python
#print out array of image file
print(img_gray.shape)
```

```python
#declare sigma and gauss2d matrix
sigma = 1.5
hsize = 2*math.ceil(3*sigma)+1
gfilt = Gauss2D((hsize,hsize),sigma)
```

```python
print(hsize)
```

```python
print(gfilt.shape)
```

```python
#declare x and y sobel filters
ysobel = (1/8)*np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
xsobel = (1/8)*np.array([[1,0,-1],[2,0,-2],[1,0,-1]])

```

```python
#run img_orig through Gauss2D filter
img_gray_blurred = myImageFilter(img_gray,gfilt)
plt.imshow(img_gray_blurred,cmap="gray");plt.show()
print(img_gray_blurred)
```

```python
#acquire x and y sobel filters from imagefilter function
Iy = myImageFilter(img_gray_blurred,ysobel)
Ix = myImageFilter(img_gray_blurred,xsobel)
```

```python
#show xsobel
plt.imshow(Ix,cmap="gray");plt.show()
print(Ix.shape)
```

```python
#show ysobel
plt.imshow(Iy,cmap="gray");plt.show()
print(Iy.shape)
```

```python
#compute magnitude matrix
Im = np.int_(np.sqrt(np.square(Ix)+np.square(Iy)))
print(Im)
```

```python
plt.imshow(Im,cmap="gray");plt.show()
print(Im.dtype)
```

```python
print(Im.shape)
print(np.amax(Im))
print(np.amin(Im))
```

```python
#comput Io orientation matrix
#np.ones_lke(Iy)*math.pi/2
#np.full(Iy.shape,math.pi/2)

Io_float = np.arctan(np.divide(Iy,Ix,out=(np.ones_like(Iy)*math.pi/2),where=Ix!=0))*(180/math.pi)
Io_float[(np.int_(np.absolute((Io_float-90)))==0) & ((Io_float) <0)] = -1*90
Io = np.int_(Io_float)
Io[Io<0] = 180 + Io[Io<0]
plt.imshow(Io,cmap="gray");plt.show()
print(Io.shape)
```

```python
print(Io.dtype)
```

```python
print(Io)
print(np.amax(Io))
print(np.amin(Io))
```
